# 와인 맛에 대한 이진 분류

```python
import pandas as pd

red_wine = pd.read_csv("../dataset/winequality-red.csv", sep=';')
white_wine = pd.read_csv("../dataset/winequality-white.csv", sep=';')

display(red_wine)
display(white_wine)
```

```python
red_wine['color'] = 1.
white_wine['color'] = 0.

wine = pd.concat([red_wine, white_wine])
wine
```

## 🔰 'quality'컬럼 이진화

- 'quality'(등급)컬럼을 'taste'(맛)으로 이진화 한다.

	- 등급이 6이상이면 1(맛있다)
	- 등급이 5이하이면 0(맛없다)

```python
wine['taste'] = [1. if grade > 5 else 0. for grade in wine['quality']]
display(wine)
wine.info()
```

#### --▶ __라벨 분할__

- 먼저 학습 데이터와 정답 데이터를 만들어 준다.

- 와인의 맛을 구별하는 것이 목적이므로 맛('taste')을 라벨로 분리해 준다.

- 라벨이 정답 데이터가 된다.

```python
X = wine.drop(['taste'], axis=1) #--> column drop
y = wine['taste']
```

#### --▶ __훈련용/테스트용 데이터로 분리__

- 데이터를 8:2 비율로 훈련용과 테스트용으로 나눠준다.

```python
from sklearn.model_selection import train_test_split
import numpy as np

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

np.unique(y_train, return_counts=True)
```

#### --▶ __훈련용 데이터로 학습__

```python
from sklearn.tree import DecisionTreeClassifier

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train) #<--(features, labels)
```

#### --▶ __학습 결과 확인__

- 정답 데이터(y_train, y_test)와 학습한 모델에 의해 예측된 데이터(y_pred_tr, y_pred_test)를 비교해서</br>
	정확도 값을 구한다.

```python
from sklearn.metrics import accuracy_score

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc:', accuracy_score(y_train, y_pred_tr)) #--> 훈련용 데이터의 정확도
print('Test Acc:', accuracy_score(y_test, y_pred_test)) #--> 테스트용 데이터의 정확도
```

👉 응? 100%? 가능한가?

- 불가능한 결과이므로 무조건 의심해야 한다.

- 왜 이런 일이 생겼는지 확인해 본 결과 모델이 'quality'컬럼으로 학습을 한 것을 볼 수 있다.

- quality 컬럼으로 taste 컬럼을 만들었으니 quality 컬럼은 제거했어야 했다.

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree

plt.figure(figsize=(7,5))
plot_tree(wine_tree, feature_names=X.columns.tolist(), fontsize=12, filled=True, rounded=True);
```

#### --▶ 재학습

- 'quality' 컬럼도 제거한 후 데이터 세팅을 다시 해준다.

```python
X = wine.drop(['taste','quality'], axis=1) #--> column drop
y = wine['taste']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=13)

wine_tree = DecisionTreeClassifier(max_depth=2, random_state=13)
wine_tree.fit(X_train, y_train) #<--(features, labels)

y_pred_tr = wine_tree.predict(X_train)
y_pred_test = wine_tree.predict(X_test)

print('Train Acc:', accuracy_score(y_train, y_pred_tr))
print('Test Acc:', accuracy_score(y_test, y_pred_test))
```

- 화학 성분으로 맛 구분을 어떻게 하는지...
