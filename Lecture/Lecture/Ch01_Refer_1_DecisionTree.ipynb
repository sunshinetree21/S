## üî∞ Iris ÌÖçÏä§Ìä∏ ÏÑ§Î™Ö(Export_Text)

- sklearn Ìå®ÌÇ§ÏßÄÏóêÎäî ÌÖçÏä§Ìä∏Î°ú Í≤∞Ï†ï Ìä∏Î¶¨Î•º ÏÑ§Î™ÖÌïòÎäî Í∏∞Îä•ÎèÑ Ï†úÍ≥µÌïúÎã§.

```python
from sklearn.datasets import load_iris
from sklearn.tree import export_text
from sklearn import tree

iris = load_iris()
X, y = iris.data, iris.target

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)

r = export_text(clf, feature_names=iris['feature_names'])
print(r)
```

## üî∞ Iris ÎÖ∏Îìú Í∑∏ÎûòÌîΩ

- ÎÖ∏Îìú Î™®ÏñëÏùÑ Í∑∏ÎûòÌîΩÏúºÎ°ú Ï∂úÎ†•ÌïòÎäî Î∞©Î≤ïÎèÑ Ï°¥Ïû¨ÌïúÎã§.

- plot_treeÏôÄ pyplot ÎùºÏù¥Î∏åÎü¨Î¶¨Î•º ÌôúÏö©ÌïòÎ©¥ Í≤∞Ï†ï Ìä∏Î¶¨Î•º Visualization Ìï† Ïàò ÏûàÍ≤å ÎêúÎã§.

```python
from sklearn.datasets import load_iris
from sklearn.tree import plot_tree
from sklearn import tree
import matplotlib.pyplot as plt

iris = load_iris()
X, y = iris.data, iris.target

clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)

plt.figure(figsize=(14,11))
plot_tree(clf, filled=True)
plt.title("Iris - Decision Tree Model")
plt.show()
```
