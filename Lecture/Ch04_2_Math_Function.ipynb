## üî∞ Î°úÍ∑∏ Ìï®Ïàò

- 2Ïùò Î™á ÏäπÏùÑ ÌïòÎ©¥ 16Ïù¥ Îê†Íπå?

	$~~~\boxed{~ 2^x = 16~}$

- Í∑∏Î†áÍ≤å Îì±Ïû•Ìïú Í≤ÉÏù¥ $\log$ Ïù¥Îã§.

	$~~~\boxed{~x=\log_2 16=4~}$

</br>

$~~~\large \text{ ‚û° Î°úÍ∑∏ Ìï®Ïàò :}~~~~\LARGE \boxed{~f(x)=\log_a x~}$

```python
# Î∞ëÏàò Î≥ÄÌôò Í≥µÏãùÏùÑ ÌôúÏö©Ìï¥ Îã§ÏñëÌïú Î∞ëÏàòÎ•º Ï∑®Ìï† Ïàò ÏûàÎèÑÎ°ù Ìï®ÏàòÎ•º Ï†ïÏùò

def log(x, base):
    return np.log(x) / np.log(base)
```

```python
x1 = np.linspace(0.0001, 5, 1000)
x2 = np.linspace(0.01, 5, 100)

y11, y12 = log(x1, 10), log(x2, np.e)
y21, y22 = log(x1, 1/10), log(x2, 1/np.e)

# y11, y12 = np.log10(x1), np.log(x2)
# y21, y22 = -np.log10(x1), -np.log(x2)

fig, ax = plt.subplots(1,2,figsize=(12,9))

ax[0].plot(x1, y11, label = '$\log_{10} x$', color='b')
ax[0].plot(x2, y12, label = '$\log_{e} x$', color='r', ls='--', lw=2)
ax[0].legend(fontsize=15, loc='lower right')
ax[0].set_xlabel('$x$', fontsize=17)
ax[0].set_ylabel('$y$', fontsize=17, rotation=0.)

ax[1].plot(x1, y21, label = '$\log_{1/10} x$', color='b')
ax[1].plot(x2, y22, label = '$\log_{1/e} x$', color='r', ls='--', lw=2)
ax[1].legend(fontsize=15, loc='upper right')
ax[1].set_xlabel('$x$', fontsize=17)
ax[1].set_ylabel('$y$', fontsize=17, rotation=0.)

plt.show()
```

## üî∞ ÏãúÍ∑∏Î™®Ïù¥Îìú Sigmoid


$~~~~~~~~~~\LARGE \boxed{\sigma(z)=\cfrac{1}{~1+e^{-z}~}}$

- e Îäî ÏûêÏó∞ÏÉÅÏàòÎ°ú ÎåÄÎûµ 2.71828Ïùò Í∞íÏùÑ Í∞ÄÏßÑÎã§.

- Î≥ÄÏàò $z$Ïóê ÎåÄÌïú Ìï®ÏàòÎ°ú,

- 0Í≥º 1 ÏÇ¨Ïù¥Ïùò Í∞íÏùÑ Í≤∞Í≥ºÎ°ú Î∞òÌôòÌïúÎã§. ‚≠ê **ÏãúÍ∑∏Î™®Ïù¥Îìú Ìï®ÏàòÎäî 0Í≥º 1Î°ú ÏàòÎ†¥ÌïúÎã§.** ‚≠ê

- Ïù¥ÏßÑ Î°úÏßÄÏä§Ìã± ÌöåÍ∑ÄÏóêÏÑú ÌôïÎ•†ÏùÑ Íµ¨Ìï† Îïå ÏÇ¨Ïö©ÎêúÎã§. Ï£ºÎ°ú 0.5Î•º Í∏∞Ï§ÄÏúºÎ°ú ÌÅ¥ÎûòÏä§Î•º Î∂ÑÎ•òÌïòÍ≤å ÎêúÎã§.

```python
z = np.linspace(-10, 10, 100)
sigmoid = 1 / (1 + np.exp(-z))

plt.figure(figsize=(12,8))
plt.plot(z, sigmoid)
plt.title('Sigmoid Function', fontsize=21)
plt.xlabel('$z$', fontsize=17)
plt.ylabel('$\sigma(z)$', fontsize=17, rotation=0.)
plt.show()
```

# Ìï®ÏàòÏùò ÌëúÌòÑ

## üî∞ Î≤°ÌÑ∞Ïùò ÌëúÌòÑ

<img src="https://github.com/ElaYJ/Study_Machine_Learning/assets/153154981/70c6f32a-eb98-412f-887e-56bbe2dde8da" width="50%" height="50%"></br>

- Î≥¥ÌÜµ Ïó¥Î≤°ÌÑ∞ ÌòïÌÉúÍ∞Ä ÎßéÎã§.

- Ï±ÖÏù¥ÎÇò ÎÖºÎ¨∏ÏóêÏÑúÎäî Transpose(T, ÌñâÍ≥º Ïó¥ ÍµêÌôò)Î•º ÌÜµÌï¥ Ïö∞Ï∏°Ï≤òÎüº ÌñâÎ≤°ÌÑ∞Î°ú ÌëúÌòÑÌïòÎäî Í≤ΩÏö∞Í∞Ä ÎßéÎã§.

</br>

### ‚è∫ Îã®Ïùº Î≥ÄÏàò Ïä§ÏπºÎùº Ìï®Ïàò

$~~~~~~~\huge \boxed{~y=f(x)~}$

- Ïä§ÏπºÎùº Í∞íÏúºÎ°ú ÏûÖÎ†•ÎêòÍ≥† Ïä§ÏπºÎùº Í∞íÏúºÎ°ú Í≤∞Í≥ºÍ∞Ä Ï∂úÎ†•ÎêòÎäî Ìï®ÏàòÏù¥Îã§.

- Î≥ÄÏàòÍ∞Ä $x$ ÌïòÎÇòÏù¥Î∏åÎ°ú Îã®Ïùº Î≥ÄÏàò~!

</br>

### ‚è∫ Îã§Ï§ë Î≥ÄÏàò Ïä§ÏπºÎùº Ìï®Ïàò

$~~~~~~~\huge \boxed{~y=f(\vec{x})~}$

- $~\vec{x} ~~\dashrightarrow ~~~x=[\cdots]~~$ xÍ∞Ä Ïó¨Îü¨ Í∞úÏùò Í∞íÏùÑ Ïó¥Î≤°ÌÑ∞Î°ú Í∞ÄÏ°åÎã§ Í≤ÉÏùÑ ÏùòÎØ∏ÌïúÎã§.

- ÏûÖÎ†•ÏùÄ Ïó¨Îü¨ Í∞úÏù¥ÎØÄÎ°ú Îã§Ï§ë Î≥ÄÏàòÏù¥Í≥† Ï∂úÎ†•Ïù¥ Ïà´Ïûê ÌïòÎÇòÏùò ÌòïÌÉúÏù¥ÎØÄÎ°ú Ïä§ÏπºÎùº Ìï®ÏàòÏù¥Îã§.

</br>

### ‚è∫ Îã§Ï§ë Î≥ÄÏàò Î≤°ÌÑ∞ Ìï®Ïàò

<img src="https://github.com/ElaYJ/Study_Machine_Learning/assets/153154981/6c10cf5c-8e3f-477c-98db-b01424199d68" width="50%" height="50%"></br>


- $~\vec{x},~~\Bbb{X}~$ Îëò Îã§ Î≤°ÌÑ∞Î•º ÎÇòÌÉÄÎÇ¥Îäî Í∏∞Ìò∏Ïù¥Îã§.

- Ïó¨Îü¨ Í∞úÏùò Ïó¥Î°ú Ï∂úÎ†•ÎêúÎã§.

- Îã§Î≥ÄÏàò Î≤°ÌÑ∞Ìï®ÏàòÎäî Ïó¨Îü¨ Í∞úÏùò ÎèÖÎ¶ΩÎ≥ÄÏàòÎ•º Í∞ÄÏßÄÍ≥† ÏûàÏúºÎ©∞, Í≤∞Í≥ºÎ°ú Î≤°ÌÑ∞ Í∞íÏùÑ Î∞òÌôòÌïòÎäî Ìï®ÏàòÏù¥Îã§.</br>
	Ï¶â, ÏûÖÎ†•Î≥ÄÏàò $x$ Ïóê ÎåÄÌï¥ Ïó¨Îü¨ Í∞úÏùò Ìï®Ïàò $f_1,~f_2,\cdots,~f_n~$Ïùò Í≤∞Í≥ºÎ•º Î≤°ÌÑ∞ ÌòïÌÉúÎ°ú Î¨∂Ïñ¥ÏÑú Î∞òÌôòÌïúÎã§.

- ÏûÖÎ†•Î≥ÄÏàòÍ∞Ä $x$ Ïù¥Í≥†, $f_1,~f_2,\cdots,~f_n~$Ïù¥ $x$ Ïùò Í∞Å ÏöîÏÜåÏóê ÎåÄÌïú Ìï®ÏàòÏùº Îïå,

	Îã§Î≥ÄÏàò Î≤°ÌÑ∞Ìï®Ïàò $F(\vec{x}) = \begin{pmatrix} f_1(\vec{x}) \\ f_2(\vec{x}) \\ \vdots \\ f_n(\vec{x}) \end{pmatrix}~$ Î°ú Ï†ïÏùòÎêúÎã§.

	- Ïó¨Í∏∞ÏÑú $F(\vec{x})$Îäî nÏ∞®Ïõê Î≤°ÌÑ∞Î•º Î∞òÌôòÌïòÎ©∞, Í∞Å ÏöîÏÜåÎäî ÏûÖÎ†•Î≥ÄÏàò $\vec{x}$ Ïóê ÎåÄÌïú Í∞ÅÍ∞ÅÏùò Ìï®Ïàò $f_i$ Ïùò Í∞íÏù¥Îã§.<br></br>

- Î™®Îì† $f_i$Ìï®ÏàòÎì§Ïù¥ Í∞ôÏùÄ ÎèÖÎ¶ΩÎ≥ÄÏàò Î≤°ÌÑ∞ $\vec{x}$ Î•º Í≥µÏú†ÌïòÏßÄÎßå, Í∞Å Ìï®ÏàòÎäî $\vec{x}$ Î•º Îã§Î•¥Í≤å Ìï¥ÏÑùÌïòÍ±∞ÎÇò Îã§Î•∏ Í≥ÑÏÇ∞ÏùÑ ÏàòÌñâÌïòÏó¨ Í∞ÅÍ∞Å Îã§Î•∏ Í≤∞Í≥ºÎ•º Î∞òÌôòÌï† Ïàò ÏûàÎã§.

- Í≤∞Í≥ºÏ†ÅÏúºÎ°ú $F(\vec{x})$Îäî $f_i(\vec{x})~$Í∞Ä Ï∂úÎ†•(Î∞òÌôò)ÌïòÎäî Í∞íÏùÑ ÏöîÏÜåÎ°ú Í∞ÄÏßÄÎäî Î≤°ÌÑ∞Î•º ÏÉùÏÑ±ÌïòÍ≤å ÎêúÎã§.

	1. **ÏûÖÎ†• Î≤°ÌÑ∞¬†$\vec{x}$**: Ïù¥Îäî¬†$m$Ï∞®Ïõê Í≥µÍ∞ÑÏùò Ï†êÏùÑ ÎÇòÌÉÄÎÇº Ïàò ÏûàÎã§. Ïó¨Í∏∞ÏÑú¬†$m$ÏùÄ¬†$\vec{x}$Ïùò ÏöîÏÜå Í∞úÏàòÏù¥Îã§.

	2. **Í∞Å Ìï®Ïàò¬†$f_i$**: Ïù¥ Ìï®ÏàòÎì§ÏùÄ¬†$\vec{x}$Ïùò Í∞Å ÎèÖÎ¶ΩÎ≥ÄÏàòÎì§ÏùÑ Î∞õÏïÑÏÑú ÌïòÎÇòÏùò Ïä§ÏπºÎùº Í∞íÏùÑ Ï∂úÎ†•ÌïúÎã§. Ïó¨Í∏∞ÏÑú¬†$i$Îäî 1Î∂ÄÌÑ∞¬†$n$ÍπåÏßÄÏùò Ïù∏Îç±Ïä§Ïù¥Îã§.

	3. **Ï∂úÎ†• Î≤°ÌÑ∞¬†$F(\vec{x})$**: Ïù¥ Î≤°ÌÑ∞Îäî¬†$n$Ï∞®Ïõê Í≥µÍ∞ÑÏùò Ï†êÏúºÎ°ú ÌëúÌòÑÎêúÎã§. Ïó¨Í∏∞ÏÑú Í∞Å¬†$f_i$Ïùò Í≤∞Í≥ºÎäî¬†$F(\vec{x})$Î≤°ÌÑ∞Ïùò¬†$i$Î≤àÏß∏ ÏöîÏÜåÍ∞Ä ÎêúÎã§.


#### üìç __`numpy.meshgrid()`__

- numpy ÎùºÏù¥Î∏åÎü¨Î¶¨ÏóêÏÑú Î≤°ÌÑ∞ x, yÏóê ÎåÄÌï¥ Î™®Îì† (x,y)ÏåçÏùÑ ÌëúÏãúÌï† Ïàò ÏûàÎäî Ìï®ÏàòÏù¥Îã§.

- Îëê Î≤°ÌÑ∞Î•º ÏûÖÎ†• Î∞õÏïÑÏÑú 2Ï∞®Ïõê Í≤©Ïûê(grid)Î•º ÌòïÏÑ±ÌïúÎã§.

- Ïù¥ Ìï®ÏàòÎäî Ï£ºÎ°ú 2Ï∞®Ïõê Í≥µÍ∞ÑÏóê ÎåÄÌïú ÌèâÍ∞Ä ÏßÄÏ†êÎì§ÏùÑ ÏÉùÏÑ±ÌïòÎäî Îç∞ ÏÇ¨Ïö©ÎêòÎ©∞, 3Ï∞®Ïõê Í≥µÍ∞ÑÏúºÎ°ú ÌôïÏû•Ìï¥ÏÑú ÏÇ¨Ïö©Ìï† ÏàòÎèÑ ÏûàÎã§.

- Í≤©ÏûêÎ•º ÌòïÏÑ±ÌïòÎäî ÏõêÎ¶¨Îäî Í∞Å ÏûÖÎ†• Î≤°ÌÑ∞Ïùò Î™®Îì† ÏöîÏÜåÎì§ Í∞ÑÏùò Ï°∞Ìï©ÏùÑ ÎßåÎì§Ïñ¥ÎÇ¥Îäî Í≤ÉÏù¥Îã§.

</br>

$\boxed{\bold{ÏõêÎ¶¨}}$

- ÏòàÎ•º Îì§Ïñ¥, Îëê Î≤°ÌÑ∞¬†$\vec{x}$ = [x1, x2, x3]Í≥º¬†$\vec{y}$ = [y1, y2, y3, y4]Í∞Ä ÏûàÎã§Í≥† Í∞ÄÏ†ïÌï¥ Î≥¥Ïûê.

- `np.meshgrid(x, y)`Îäî Îã§ÏùåÍ≥º Í∞ôÏù¥ ÏûëÎèôÌïúÎã§:

	1. $\vec{x}$¬†Î≤°ÌÑ∞Îäî Í∞Å¬†$\vec{y}$¬†ÏöîÏÜåÏóê ÎåÄÌï¥ Î∞òÎ≥µÎêúÎã§. Ïù¥Í≤ÉÏùÄ 2Ï∞®Ïõê Î∞∞Ïó¥ÏóêÏÑú Í∞Å ÌñâÏù¥¬†$\vec{x}$¬†Î≤°ÌÑ∞Ïùò Î≥µÏÇ¨Î≥∏Ïù¥ ÎêòÎèÑÎ°ù ÌïúÎã§.
	2. $\vec{y}$¬†Î≤°ÌÑ∞Îäî Í∞Å¬†$\vec{x}$¬†ÏöîÏÜåÏóê ÎåÄÌï¥ Î∞òÎ≥µÎêúÎã§. Ïù¥Í≤ÉÏùÄ 2Ï∞®Ïõê Î∞∞Ïó¥ÏóêÏÑú Í∞Å Ïó¥Ïù¥¬†$\vec{y}$¬†Î≤°ÌÑ∞Ïùò Î≥µÏÇ¨Î≥∏Ïù¥ ÎêòÎèÑÎ°ù ÌïúÎã§.

	3. Í≤∞Í≥ºÏ†ÅÏúºÎ°ú, Îëê Í∞úÏùò 2Ï∞®Ïõê Î∞∞Ïó¥Ïù¥ ÏÉùÏÑ±ÎêúÎã§. ÌïòÎÇòÎäî¬†$\vec{x}$¬†Ï¢åÌëúÎ•º ÏúÑÌïú Í≤ÉÏù¥Í≥†, Îã§Î•∏ ÌïòÎÇòÎäî¬†$\vec{y}$¬†Ï¢åÌëúÎ•º ÏúÑÌïú Í≤ÉÏù¥Îã§.

- Ïù¥ Îëê Î∞∞Ïó¥ÏùÑ Ìï®Íªò ÏÇ¨Ïö©ÌïòÏó¨ 2Ï∞®Ïõê Í≥µÍ∞ÑÏóêÏÑú Î™®Îì†¬†`(x, y)`¬†ÏåçÏùÑ ÎÇòÌÉÄÎÇº Ïàò ÏûàÎã§.

- `np.meshgrid`Ïùò Í≤∞Í≥ºÎäî ÏûÖÎ†•Îêú Î∞∞Ïó¥Ïùò Í∏∏Ïù¥Ïóê Í∏∞Î∞òÌïòÏó¨ ÏÉùÏÑ±ÎêòÎäîÎç∞, Í∞ÅÍ∞ÅÏùò Ï∂úÎ†• Î∞∞Ïó¥Ïùò Ï∞®ÏõêÏùÄ ÏûÖÎ†• Î∞∞Ïó¥Îì§Ïùò Í∏∏Ïù¥Ïóê ÏùòÌï¥ Í≤∞Ï†ïÎêúÎã§. 

- Ï¶â,¬†$\vec{x}$ÏôÄ¬†$\vec{y}$¬†Î∞∞Ïó¥Ïù¥ Ï£ºÏñ¥Ïßà Îïå,¬†$\vec{x}$Ïùò Í∏∏Ïù¥Í∞Ä¬†$n$Ïù¥Í≥†¬†$\vec{y}$Ïùò Í∏∏Ïù¥Í∞Ä¬†$m$Ïù¥ÎùºÎ©¥,¬†</br>
	`np.meshgrid(x, y)`Î•º ÌÜµÌï¥ ÏÉùÏÑ±ÎêòÎäî Îëê Í∞úÏùò 2Ï∞®Ïõê Î∞∞Ïó¥¬†$\vec{x},~~\vec{x}$Îäî Í∞ÅÍ∞Å¬†`m x n`¬†Ï∞®ÏõêÏùÑ Í∞ÄÏßÄÍ≤å ÎêúÎã§.

```python
t = np.linspace(0, 4, 5)
p = np.linspace(0, 4, 5)
print("t:", t)
print("p:", p)
```

#### --‚ñ∂ Îã§Î≥ÄÏàò Î≤°ÌÑ∞ Ìï®Ïàò ÏòàÏ†ú

- Î≥ÄÏàò 2Í∞ú ÏûÖÎ†•, Í≤∞Í≥º 3Í∞ú Ï∂úÎ†•

```python
u = np.linspace(0, 1, 30)
v = np.linspace(0, 1, 30)

U, V = np.meshgrid(u, v)

X = U
Y = V
Z = (1 + U**2) + (V / (1 + V**2))
```

```python
x = np.linspace(-4,4,100)

y = x**3 - 15*x + 30 #--> f(x)
z = np.log(y) #--> g(x)
```

- ÏùºÎã® Ìï®Ïàò Í∞ÅÍ∞ÅÏùÑ Îî∞Î°ú ÏãúÍ∞ÅÌôî

```python
fig, ax = plt.subplots(1,2,figsize=(12,6))

ax[0].plot(x, y, label = '$x^3 - 15x + 30$', color = 'b')  #, ls = ':'
ax[0].legend(fontsize=17)

ax[1].plot(y, z, label = '$\log(y)$')
ax[1].legend(fontsize=17)

plt.show()
```

- Îëê Ìï®Ïàò Ìï©ÏÑ±Ìï¥ÏÑú ÏãúÍ∞ÅÌôî

	$~\large \boxed{g \circ f(x)} = g(f(x)) = \log(x^3-15x+33)$

```python
fig, ax = plt.subplots(1,2,figsize=(12,6))

ax[0].plot(x, z, label = '$g(f(x))=\log(f(x))$', color = 'b')
ax[0].legend(fontsize=17)

ax[1].plot(x, y, label = '$f(x)=x^3 - 15x + 30$', color = 'k', ls=':')
ax[1].legend(fontsize=17)

plt.show()
```
